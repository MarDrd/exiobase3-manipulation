# Régions - EXIOBASE3

Ce code utilise la librairie 'country_converter' pour convertir les codes des régions provenant de EXIOBASE3 en leurs noms officiels et en leurs continents respectifs. Cela permet également de visualiser tous les pays regroupés dans les régions identifiées par les codes EXIOBASE3 'WA', 'WE', 'WF', 'WL' et 'WM'.

Pour installer 'country_converter':

pip install country_converter --upgrade


Vous pouvez trouver plus d'informations sur la librairie 'country_converter' en cliquant [ici](https://github.com/IndEcol/country_converter).

```python
import pandas as pd
import country_converter as coco
```

```python
#Définir les noms des régions et des continents associés à chaque code EXIO3
converter = coco.CountryConverter()

#Liste des codes de région EXIOBASE3
exio3_region = ['AT', 'AU', 'BE', 'BG', 'BR', 'CA', 'CH', 'CN', 'CY', 'CZ', 'DE', 'DK', 'EE', 'ES', 'FI', 'FR', 'GB', 'GR', 'HR', 'HU', 'ID', 'IE', 'IN', 'IT', 'JP', 'KR', 'LT', 'LU', 'LV', 'MT', 'MX', 'NL', 'NO', 'PL', 'PT', 'RO', 'RU', 'SE', 'SI', 'SK', 'TR', 'TW', 'US', 'WA', 'WE', 'WF', 'WL', 'WM', 'ZA']

#Conversion des codes EXIOBASE3 en noms officiels
df = converter.convert(names=exio3_region, src="EXIO3", to="name_official")

#Conversion des codes EXIOBASE3 avec le nom du continent auquel les régions appartiennent
df1 = converter.convert(names=exio3_region, src="EXIO3", to="continent")

#Convertir en DataFrame
df = pd.DataFrame(df)
df1 = pd.DataFrame(df1)

#Création de la DataFrame avec le code des régions 
df2 = pd.DataFrame(exio3_region)

#Nommer le nom de la colonne
df.columns = ['Région']
df1.columns = ['Continent']
df2.columns = ['Code EXIO3']

#Création d'une nouvelle DataFrame en mergeant les 3 DataFrame
merge = pd.concat([df2, df, df1], axis=1)

#Permet d'afficher les cellules entières de la DataFrame
pd.set_option('display.max_colwidth', None)

#Afficher la DataFrame
merge
```
